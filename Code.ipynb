```python
#Importing Packages
import matplotlib.pyplot as plt
```

```python
#Regression Function
def linear_regression(x,y):
  n = len(x)
  x_sum = sum(x)
  y_sum = sum(y)
  xy_sum = sum([xi*yi for xi,yi in zip(x,y)])
  x_squared_sum = sum([xi**2 for xi in x])

  #Calculate b1(or slope) and b0(or intercept)
  b1 = (n*xy_sum - x_sum*y_sum)/(n*x_squared_sum - x_sum**2)
  b0 = (y_sum-b1*x_sum)/n

  return b1, b0
```

```python
#Driver Code
x = [5,6,8,10]
y = [11,11,15,24]

b1, b0 = linear_regression(x,y)
```

```python
print("Slope -",b1)
```

```python
print("Intercept -",b0)
```

```python
#Calc Predicted Values
pred_line_values = [b1 * i + b0 for i in x]
```

```python
#Plotting Graph
plt.scatter(x,y)
plt.plot(x, pred_line_values, 'r')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Simple Linear Regression')
plt.show()
```
